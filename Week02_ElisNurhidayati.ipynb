## 1. Hasil citra dengan channel warna BGR dan mengetahui ukuran panjang dan lebar dengan plot matplotlib

```python
import cv2 as cv
#from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/Elis - 3C/SMT 5/PCVK_image/pororo.jfif')
plt.imshow(img)   # Hasilnya adalah citra dengan channel warna BGR
```

## 2. OpenCV membaca image dan menyimpan dalam channel warna BGR (Blue Green Red)

```python
img2 = img.copy()
img3 = cv.cvtColor(img2, cv.COLOR_BGR2RGB) # conversi BGR -> RGB
plt.imshow(img3)
```

## 3. Menampilkan citra Grayscale, melakukan resizing, melakukan flipping, menyimpan citra hasil

```python
img_gray = cv.imread('/content/drive/MyDrive/Elis - 3C/SMT 5/PCVK_image/pororo.jfif', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

**Colormap dengan warna 'gray'**

```python
plt.imshow(img_gray, cmap='gray')
```

**Colormap dengn warna 'magma'**

```python
plt.imshow(img_gray, cmap='magma')
```

**Citra RGB di resize dalam ukuran panjang 512 dan lebar 1024**

```python
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
```

**Citra RGB ditampilkan dalam ukuran yang lebih besar dengan posisi gambar terbalik**

```python
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB),0)

# tampilkan plot dengan ukuran canvas yang lebih besar
fig = plt.figure(figsize=(10,10))  # ubah-ubah ukuran (10, 10) sesuai kebutuhan
ax = fig.add_subplot(111)
ax.imshow(img5)
```

## 4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16

```python
black_img = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

**Menambahkan bentuk persegi di sebelah kanan atas sesuai koordinan pt1 dan pt2**

```python
cv.rectangle(black_img, pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black_img)
```

**Menambahkan bentuk persegi di tengah sesuai koordinan pt1 dan pt2**

```python
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

**Menambahkan bentuk lingkarang sesuai radius**

```python
cv.circle(black_img, center=(100,100), radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

**Menambahkan garis diagonal sesuai koordinat pt1 dan pt2**

```python
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

**Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan**

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img,text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

**Pembuatan black image kembali dilakukan dengan tipe data int32**

```python
black_img2=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

**inisialisasi NumPy array dengan tipe data int32**

```python
vertices = np.array([[100,300], [200,200],[400,300],[200,400]], dtype=np.int32)
vertices
```

**Reshape array**

```python
pts = vertices.reshape((-1,1,2))  # nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili R, G, B
pts
```

**Penambahan polyline pada black image kedua yang telah dibuat**

```python
cv.polylines(black_img2, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black_img2)
```

1. **Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?**

**Jawab:** Matplotlib adalah library Python yang digunakan untuk visualisasi data, seperti membuat grafik atau plot. Grafik yang dihasilkan oleh Matplotlib terdiri dari satu atau lebih sumbu koordinat, dengan sumbu horizontal (x) dan sumbu vertikal (y). Data ditampilkan menggunakan warna dan elemen visual seperti marker (misalnya, lingkaran), garis, atau poligon. Tanpa Matplotlib, grafik tidak akan muncul. Gambar yang ditampilkan melalui OpenCV menggunakan format BGR, sedangkan dengan Matplotlib, gambar menggunakan format RGB, yang lebih cocok untuk pengolahan citra. Selain itu, Matplotlib menyediakan fitur tambahan seperti anotasi dan colorbars untuk memperkaya visualisasi data.

**Jawab:** Tipe data int16 dan int32 berbeda dalam penggunaan memori dan kecepatan, dengan int32 memakan lebih banyak memori dan lebih lambat. Meskipun keduanya sering berlebihan untuk gambar dengan rentang piksel 0-255, gambar hitam yang dibuat dengan int32 akan menghasilkan file yang lebih besar dibandingkan dengan int16, mempengaruhi intensitas dan ukuran file gambar.

8. **Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB dari suatu citra!**
**Jawab:**

```python
img_sbob = cv.imread('/content/drive/MyDrive/Elis - 3C/SMT 5/PCVK_image/sbob.jfif')
img_sbob = cv.cvtColor(img_sbob, cv.COLOR_BGR2RGB)
# Memperoleh dimensi citra
height, width, channels = img_sbob.shape

# Menentukan ukuran dan warna rectangle
rect_width = 100  # Lebar rectangle dalam piksel
rect_height = 100  # Tinggi rectangle dalam piksel
rect_color = (0, 0, 255)  # Warna biru (BGR) untuk rectangle

# Menentukan koordinat sudut bawah kiri rectangle
x1 = 0  # Koordinat x
y1 = height - rect_height  # Koordinat y (sudut bawah)

# Menggambar rectangle pada channel B (biru) dari citra RGB
img_sbob[y1:y1+rect_height, x1:x1+rect_width, 0] = rect_color[0]  # Channel B

# Menyimpan citra yang telah diubah
plt.imshow(img_sbob)
```
